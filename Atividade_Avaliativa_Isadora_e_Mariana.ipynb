**Quest√£o 2**

```python
n = 10 # numero de pontos
g = 9.8
y = 2
m = 10

t = np.linspace(0,10,n)
v = np.linspace(40,60,n)

T,V = np.meshgrid(t,v)

a = g - y/m*V
ang = np.arctan(a) 
dv = np.cos(ang)
dt = np.sin(ang)

plt.quiver(T,V,dv,dt,angles='xy')
plt.grid('on')
```

```python
# funcao para calcular a acelera√ß√£o (equa√ß√£o diferencial)
def ed(v,t):
  a = g-y/m*v
  return a

tt = np.linspace(0,10,10)
# odeint = resolve um sistema de equa√ßoes diferenciais - parametros: a ED (ed), condicao inicial em y(y0), sequ√™ncia de pontos de tempo (tt)
y0 = 40
sol = odeint(ed,y0,tt)
plt.plot(tt,sol) # plota a curva
plt.quiver(T,V,dv,dt)
plt.grid('on')
```

```python
plt.figure(figsize = (6,6),dpi=100)
plt.quiver(T,V,dv,dt,angles='xy',label='$dv/dt$')
plt.grid('on')
for v0 in range(40,50,2):
  sol = odeint(ed,v0, tt)
  plt.plot(tt,sol, label=v0)
plt.legend()
plt.xlabel('t')
plt.ylabel('v')
plt.show()
```
